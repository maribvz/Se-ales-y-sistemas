Using OpenCV to perform arithmetic operations on images

```python
# Importo algunas librerìas necesarias
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

a. Suma de imágenes

```python
#Para la imagen 1
# Leo las imágenes
img1 = cv2.imread('/home/maribel/Descargas/Maribel Vallejo Zabala - water- waste1.jpg')


# Cambio escala de colores de la imagen
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)


# Guardo las imágenes
cv2.imwrite('img1.jpg', img1)


#Muestro las imágenes
plt.imshow(img1)


# Visualizo el tamaño de imagen
img1.shape
```

```python
#Para la imagen 2
# Leo las imágenes

img2 = cv2.imread('/home/maribel/Descargas/Maribel Vallejo Zabala - water- waste2.jpg')

# Cambio escala de colores de la imagen

img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

# Guardo las imágenes

cv2.imwrite('img2.jpg', img2)

#Muestro las imágenes

plt.imshow(img2)

# Visualizo el tamaño de imagen

img2.shape
```

```python
# Cambio el tamaño de la imagen 1 para poder operarla con la imagen 2
img1 = cv2.resize(img1,(728,489))
plt.imshow(img1)
img1.shape
```

```python
# Sumo las imágenes de la forma 1 que aparece en la página web
suma = img1 + img2

#Guardo la imagen suma
cv2.imwrite('suma.jpg', suma)

#muestro la imagen suma
plt.imshow(suma)
```

```python
# Sumo las imágenes de la forma 2 que aparece en la página web
suma2 = cv2.add(img1, img2)

#Guardo la imagen suma2
cv2.imwrite('suma2.jpg', suma2)

#muestro la imagen suma
plt.imshow(suma2)
```

b. Resta de imágenes

```python
# Resto las imágenes 
resta = cv2.subtract(img1, img2)
 
# Guardo la imagen resta
cv2.imwrite('resta.jpg', resta)

#muestro la imagen resta
plt.imshow(resta)
```

c. División de imágenes

```python
 # Divido la imágen1 entre una constante
division = cv2.divide(img1, 3)
 
# Guardo la imagen division
cv2.imwrite('division.jpg', division)

#muestro la imagen division
plt.imshow(division)
```

```python
# Divido la imagen 1 entre una constante con Numpy
 
# Divido con Numpy
division2 = img1 / 500


# Guardo la imagen division2
cv2.imwrite('division2.jpg', division2)

#Pongo la imagen en rango adecuado para RGB y cambio la escala de colores
division2 = np.clip(division2, 0, 1)

#muestro la imagen division2
plt.imshow(division2)
```

d. multiplicación de imágenes

```python
# Multiplico la imagen1 por una cosntante
multiplicacion = img1 * 0.004
 
# Guardo la imagen multiplicacion
cv2.imwrite('multiplicacion.jpg', multiplicacion)

#Pongo la imagen en rango adecuado para RGB 
multiplicacion = np.clip(multiplicacion, 0, 1)

#muestro la imagen multiplicacion
plt.imshow(multiplicacion)
```

```python
# Multiplico la imagen1 por la imagen2
multiplicacion2 = img1 * img2
 
# Guardo la imagen multiplicacion2
cv2.imwrite('multiplicacion2.jpg', multiplicacion2)

#Pongo la imagen en rango adecuado para RGB 
multiplicacion2 = np.clip(multiplicacion2, 0, 1)

#muestro la imagen multiplicacion2
plt.imshow(multiplicacion2)
```

```python
# Multiplico la imagen1 por una constante
multiplicacion3 = cv2.multiply(img1, 1.5)
 
# Saving the output image
cv2.imwrite('multiplicacion3.jpg', multiplicacion3)

#muestro la imagen multiplicacion3
plt.imshow(multiplicacion3)
```

e. Bitwise

Bitwise_and

```python

#cv2.bitwise_and aplicado a las imagenes
dest_and = cv2.bitwise_and(img2, img1, mask = None)
 
# Guardo la imagen Bitwise and
cv2.imwrite('bitwiseand.jpg', dest_and)

#muestro la imagen bitwise and
plt.imshow(dest_and)
```

```python
# cv2.bitwise_or aplicado a las imagenes
dest_or = cv2.bitwise_or(img2, img1, mask = None)
 
# Guardo la imagen Bitwise or
cv2.imwrite('bitwiseor.jpg', dest_or)

#muestro la imagen bitwise or
plt.imshow(dest_or)
```

```python
# cv2.bitwise_xor aplicado a las imagenes
dest_xor = cv2.bitwise_xor(img1, img2, mask = None)

# Guardo la imagen Bitwise xor
cv2.imwrite('bitwisexor.jpg', dest_xor)

#muestro la imagen bitwise xor
plt.imshow(dest_xor)
```

```python
# cv2.bitwise_not aplicado a la imagen 1
dest_not1 = cv2.bitwise_not(img1, mask = None)

# Guardo la imagen Bitwise not1
cv2.imwrite('bitwisenot1.jpg', dest_not1)

#muestro la imagen bitwise not1
plt.imshow(dest_not1)
```

```python
# cv2.bitwise_not aplicado a la imagen 2
dest_not2 = cv2.bitwise_not(img2, mask = None)

# Guardo la imagen Bitwise not2
cv2.imwrite('bitwisenot2.jpg', dest_not2)

#muestro la imagen bitwise not2
plt.imshow(dest_not2)
```
